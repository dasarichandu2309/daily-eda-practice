## Descriptive Statistics

```python
import statistics as st

marks = [50,60,60,70,80]

mean_value = st.mean(marks)
median_value = st.median(marks)
mode_value = st.mode(marks)

print("Marks:",marks)
print("Mean:" , mean_value)
print("Median:", median_value)
print("Mode:" , mode_value)
```
