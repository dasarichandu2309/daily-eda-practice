## Measure of Dispersion

```python
import statistics as st

data = [48,49,50,51,52]

#Range
data_range = max(data) - min(data)
print("Range:" , data_range)

#Population variance and standard deviation
pop_variance = st.pvariance(data)
pop_std = st.pstdev(data)

print("Population variance:" , pop_variance)
print("standard deviation:" , pop_std)

#sample variance and standard deviation
sample_variance = st.variance(data)
sample_std = st.stdev(data)

print("sample variance" , sample_variance)
print("sample standard deviation:" , sample_std)
```

### Quartiles


```python
data = [5,10,15,20,25,30,35,40,45,50,55,60]
data.sort()

#Median
def median(lst):
    n = len(lst)
    mid = n//2
    if n%2==0:
        return (lst[mid - 1] + lst[mid])/2
    else:
        return lst[mid]

Q2 = median(data)

lower = data[:len(data)//2] #Q1
upper = data[len(data)//2:] #Q3

Q1 = median(lower)
Q3 = median(upper)

IQR = Q3 - Q1

Q0 = min(data)
Q4 = max(data)

print("Q1:" , Q1)
print("Q2:" , Q2)
print("Q3:" , Q3)
print("IQR:"  , IQR)
print("Q0:" , Q0)
print("Q4:" , Q4)
```
