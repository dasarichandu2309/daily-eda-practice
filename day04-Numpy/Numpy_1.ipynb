```python
#1D array (single Dimesnison) ----> 1 index
# Vector ----> One Dimensinal array
a = np.array([1,2,3,4,5])
print(a)
print(a.ndim)
print("Shape" , a.shape) # No of elements
print("Size" , a.size) # No of elments
```

```python
# 2d array(Matrix) ---> 2 indices
b = np.array([[1,2,3,4],
              [5,6,7,8]])
print(b)
print(b.ndim)
print("Shape" , b.shape)
print("Size" , b.size)
```

```python
# 3d array ---> block ,row,column
# Cube or Tensor ---> Three Dimensinal array
c = np.array([[[1,2],
               [3,4]],

              [[5,6],
               [7,8]]])
print(c)
print(c.ndim)
print("Shape" , c.shape) # 2blocks 2 row 2 columns
print("Size" , c.size)
```

## Numpy Built in Functions

```python
# np.zeros() ---> Creates an array filled with 0s
# np.ones() ----> Creats an array filled with 1s
# np.arrange() ---> Creats an array with range of numbers
# np.linspace() ---> Creats an array with evenly spaced numbers
# np.full() ---> Creats an array with specific value
# np.eye() ---> Creats a squared indentiy matrix
# np.diag() ----> Creates a diagnal arrat from a list or extrcst diagnal
```

```python
print("Zeros:\n" , np.zeros((2,3)))
print("Ones:\n" , np.ones((2,3)))
print("Arange:\n",np.arange(0,11,2))
print("Linspace:\n" , np.linspace(0,1,5))
print("Full:\n",np.full((2,2),9)) 
print("Eye:\n" , np.eye(3))
print("Diagnol:\n" , np.diag([1,2,3]))
```

## dtype

#### int


```python
arr = np.array([-1,-2,-3] , dtype = np.int32)
print(arr)
print(arr.dtype)
```

```python
arr = np.array([1,2,3] , dtype = np.uint32)
print(arr)
print(arr.dtype)
```

#### Float

```python
arr = np.array([1.5,2.5,3.5],dtype = np.float64)
print(arr)
print(arr.dtype)
```

#### Complex

```python
# complex64  , complex128
arr = np.array([1+2j , 2+3j, 3+5j],dtype = np.complex64)
print(arr)
print(arr.dtype)
```

#### Boolean(bool)

```python
#stores True/False

arr = np.array([True , False , True] , dtype = bool)
print(arr)
print(arr.dtype)
```

#### String / Unicode

```python
# stores text values
# dtype = "U1" ----> Unicode string per element
arr = np.array(['a','b','c'] , dtype = 'U1')
print(arr)
print(arr.dtype)
```
