## astype()

```python
import numpy as np
```

```python
arr = np.array([1,2,3,4])
arr_1 = arr.astype(np.float32)
print("Original Array:" , arr,"----->",arr.dtype)
print("Converted Array:", arr_1,"----->",arr_1.dtype)
```

## Random Numbers

```python
#  np.random.rand() : generates random nubebers from 0 to 1

print(np.random.rand())
print("------------------------------")
# 1d array of 5 random element
print(np.random.rand(5))
print("------------------------------")
 #2d array(3rows ,2 colums)
print(np.random.rand(3,2))
```

```python
# np.random.randn()

print(np.random.randn())
print(np.random.randn())
print("------------------------------")
# 1d array of 5 random element
print(np.random.randn(5))
print("------------------------------")
 #2d array(3rows ,2 colums)
print(np.random.randn(3,2))
```

```python
# np.random.randint(start,stop)

a = np.random.randint(1,10,size = 5) # generates between 1 to 9 , no of elements : 5
print(a)

b = np.random.randint(1,10,size = (2,3))  # 2d array
print(b)

```

```python
# np.random.unoform()
a = np.random.uniform(1,10,size = 5) # generates between 1 to 9 , no of elements : 5
print(a)

b = np.random.uniform(1,10,size = (2,3))  # 2d array
print(b)

```

## Indexing

```python
## 1D Array
arr = np.array([10,20,30,40,50])
print("Array:", arr)
print("-" * 20)
print(arr[0])
print(arr[1])
print(arr[2])
print(arr[3])
print(arr[4])
print("-" * 20)
print(arr[-1])
print(arr[-1])
print(arr[-2])
print(arr[-3])
print(arr[-4])
```

## Slicing

```python
# Slicing : extracting a portion(sub_Array) from the array using the format
arr = np.array([10,20,30,40,50])
print("Array:", arr)
print("------------------------")
print(arr[0:4])
print(arr[0:6:2])
print(arr[::-1])
```

```python
## 2D Array Indexing
arr  = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,80]])
print("2D Array:\n" , arr)
print("------------------------")
print(arr[0,0])
print(arr[0,1])
print(arr[0,2])
print(arr[1,0])
print(arr[1,0])
print(arr[1,1])
print(arr[1,2])
print(arr[2,0])
print(arr[2,1])
print(arr[2,2])
print("-----------------------")
print(arr[-1,-1])
print(arr[-1,-2])
print(arr[-1,-3])
print(arr[-2,-1])
print(arr[-2,-1])
print(arr[-2,-3])
```

```python
# Slicing
arr  = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,80]])
print("2D Array:\n" , arr)
print("----------------------")
print(arr[0, :]) # extract first row
print(arr[1, :]) # extract second row
print(arr[2, :]) # extract third row
print("---------------------------")
print(arr[: , 0])
print(arr[: , 1])
print(arr[: , 2])
print("----------------------------")
print(arr[0:2 , :])
print("-----------------")
print(arr[0:2 , 0:2])
print("-----------------------")
print(arr[1:3 , :])
```

## Boolean indexing

```python
arr = np.array([1,2,3,4,5,6])

condition = arr > 3
print("Condition (Boolean Array):" , condition)
result = arr[condition]
print("Result"  ,result)
```

```python
arr = np.array([1,2,3,4,5,6])

print(arr[arr > 3])
```

```python
# 2D Array
arr2d  = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,80]])
print("2D Array:\n" , arr)
print("-------------------")

print("Condition: \n" ,arr2d > 50)
print("Result:" , arr2d[arr2d > 50])

```

## Updating Numpy Arrays

```python
arr = np.array([10,20,30,40,50])

# Updating single element
arr[0] = 100
print(arr)

#Updating multiple elements
arr[1:3] = 200 , 300
print(arr)

#fancy indexing
arr[[3,4]] = [400 , 500]

# updating using boolean indexing
arr[ arr > 400] = 999
print(arr)
```

```python
# Updating 2d Array

arr2d  = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,80]])
print("2D Array:\n" , arr)
print("-------------------")
# Update single element
arr2d[0,0] = 100
print(arr2d)
print("--------------------")
arr2d[1]  = [400,500,600]
print(arr2d)
print("---------------------")
arr2d[0:2, 0:2] = [[111,222],
                   [333,444]]
print(arr2d)
print("-------------------------")
arr2d[: , 2] = [333,666,777]
print(arr2d)

#fancy indexing
arr2d[[2,2],[0,1]] = [777,888]
print(arr2d)

# boolean indexing
arr2d[arr2d > 800] = 9999
print(arr2d)
print("-------------------------")
```

```python
# negative indexing
arr2d  = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,80]])
arr2d[-2:,-2:] = [[500,600],
                  [700,800]]
print(arr2d)
```

```python
arr2d  = np.array([[10,20,30],
                 [40,50,60],
                 [70,80,80]])
print(arr2d[-3:-1,-3:-1])
arr2d [-3:-1,-3:-1] = [[300,800],
                       [444,888]]
print(arr2d)
```
