```python
import numpy as np
```

```python
# Logical operator

a = np.array([10,20,30,40])

print(np.logical_and(a>15,a<35))
print(np.logical_or(a<5,a>35))
print(np.logical_not(a>25))
```

## Assignment opretor

```python
# Assignment opretor
a = np.array([10,20,30])
a+=5
print(a)
a-=5
print(a)
a*=2
print(a)
a//=2
print(a)
a%=4
print(a)
a**=2
print(a)
```

#### sin,cos

```python
angles_deg = np.array([0,20,45,90])

#converting degrees to radius 
angle_rad = np.deg2rad(angles_deg)

print("Degrees:" , angles_deg)
print("Radius:" , angle_rad)
# sin values
sin_values = np.sin(angle_rad)
print("Sin values:" , sin_values)
```

```python
angles_deg = np.array([[0,30],
                     [45,90]])

#converting degrees to radius 
angle_rad = np.deg2rad(angles_deg)

print("Degrees:" , angles_deg)
print("Radius:" , angle_rad)
# sin values
sin_values = np.sin(angle_rad)
print("Sin values:" , sin_values)
cos_values = np.cos(angle_rad)
print("cos values:" , cos_values)
```

```python
angles = np.array([0,np.pi/4, np.pi/3,np.pi/2])
cos_values = np.cos(angles)
print("Angles radiance:" , angles)
print("Cos values:" , cos_values)+
```
