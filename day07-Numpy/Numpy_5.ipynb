### mean()

```python
import numpy as np
```

```python
arr = np.array([50,70,30,90,60])
average = np.mean(arr)
print("Mean Value:" , average)
```

```python
arr2d = np.array([[1,2,3],
                  [4,5,6]])
print("Mean of all elements:" , np.mean(arr2d))
print("Mean of all columns:" , np.mean(arr2d , axis = 0))
print("Mean of all columns:" , np.mean(arr2d , axis = 1))
```

## median()

```python
arr = np.array([50,70,30,90,60])
med = np.median(arr)
print("Median Value:" ,med)
```

```python
arr2d = np.array([[1,2,3],
                  [4,5,6]])
med = np.median(arr2d)
print("Median of all elements:" , np.median(arr2d))
print("Median of all columns:" , np.median(arr2d , axis = 0))
print("Median of all columns:" , np.median(arr2d , axis = 1))
```

## variance()


```python
marks = np.array([50,70,30,90,60])
print("Variance:" , np.var(marks))
print("Standard deviation:" , np.std(marks))
```

```python
arr2d = np.array([[1,2,3],
                  [4,5,6]])

print("Variance of entire array:" , np.var(arr2d))
print("Variance of row wise:" , np.var(arr2d , axis = 1))
print("Variance of column wise:" , np.var(arr2d , axis = 0))
```

## Standard deviation

```python
marks = np.array([50,70,30,90,60])
print("Std dev:" , np.std(marks))
```

```python
arr2d = np.array([[1,2,3],
                  [4,5,6]])

print("Std dev of entire array:" , np.std(arr2d))
print("Std dev  of row wise:" , np.std(arr2d , axis = 1))
print("Std dev of column wise:" , np.std(arr2d , axis = 0))
```

### co-relation co-efficient

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

corr = np.corrcoef(x,y)
print(corr)

print("-----------------------")

x1 = [1,2,3,4,5]
y1 = [10,8,6,4,2]
corr1 = np.corrcoef(x1,y1)
print(corr1)
```

## Array Manipulation

#### concatenate()

```python
a = np.array([1,2,3])
b = np.array([4,5])
result = np.concatenate((a,b))
print(result)
```

```python
# 2D Array

a =  np.array([[1,2],
               [3,4]])
b = np.array([[5,6],
              [7,8]])
#Row Wise
row_concat = np.concatenate((a,b) , axis = 0)
print("Row Wise :\n" , row_concat)
print("----------------------------")

# Column Wise
column_concat = np.concatenate((a,b),axis  = 1)
print("Column Wise:\n" , column_concat)
print("----------------------------")
```

## Sorting Arrays

```python
arr = np.array([5,1,4,3,2])
sorted_arr = np.sort(arr)
print("Original array:" , arr)
print("Sorted array:", sorted_arr)
print("Descending array:" , sorted_arr[::-1])
```

```python
#2D Array

arr2d = np.array([[3,1,8],
                  [6,2,4]])
sorted_arr2d = np.sort(arr2d)
print("Original array:" , arr2d)
print("Sorted array:", sorted_arr2d)
print("Descending array:" , sorted_arr2d[::-1])
```

### argsort()

```python
marks  =  np.array([40,10,30,20])  #Index positions : [0,1,2,3]

sorted_indices = np.argsort(marks)
sorted_values = np.sort(marks)

print("Original Array:", marks)
print("Sort Values:" , sorted_values)
print("Sorted Indices:" , sorted_indices)
```

```python
#2D array

arr = np.array([[30,10,20],
                [80,60,40]])

print("Argsort columns:\n" , np.argsort(arr,axis = 0))
print("Argsort rows:\n" , np.argsort(arr , axis = 1))
```
