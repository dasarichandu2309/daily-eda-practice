### Series

```python
data  = [10,20,30,40,50]
s = pd.Series(data)
print(s)
```

```python
s1 = pd.Series([10,20,30] , index = ['a','b','c'])
print(s1)
```

```python
marks = pd.Series([67,56,89,75,98], index = ['john','charlie','james','ram','raj'])
print(marks)
```

```python
# Acess by label
print(marks['james'])
```

```python
# Acsess by position
print(marks[2])
```

```python
# use slicing
print(marks[0:3])
```

```python
# Vectorized operation
print(marks + 5)
```

```python
## Aggregate function ---> sum() , min(), max() ,mean() , count()
print(marks.sum())
print(marks.mean())
print(marks.min())
print(marks.max())
print(marks.count())
```

```python
# Membership

print("Chandu" in marks)
print("chandu"  not in marks)
```

```python
#Filtering
print(marks[marks > 70])
```

```python
marks['raj'] = 90
print(marks)
```

```python
# get index and values
print(marks.index)
print(marks.values)
```

```python
print(marks.dtype)
print(marks.shape)
print(len(marks))
print(marks.head()) # return top records/rows
print("-" * 30)
print(marks.tail())
print(marks.median())
print(marks.unique())
print(marks.std())
```
