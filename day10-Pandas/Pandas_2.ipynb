```python
import pandas as pd
import numpy as np
```

```python
# Create a series from dictionary
#Keys ---> index
#values ---> data

data = {"A":85 , "B":92 , "C":78 , "D":90, "E":88}
marks = pd.Series(data)
print(marks)
```

```python
#Sort a series

data = {"A":85 , "D":90 , "C":78 ,"B":92 , "E":88}
marks = pd.Series(data)
print(marks)
print("-" * 20)
#Sort by index
print(marks.sort_index())
print("-" * 20)
# Sort by values
print(marks.sort_values()) # ascending

print(marks.sort_values(ascending = False))
```

```python
# Reordering index in series
data = {"A":85 , "D":90 , "C":78 ,"B":92 , "E":88}
marks = pd.Series(data)
print(marks)
print("-" * 30)

marks = pd.Series(data,index = ["A","B","C","D","E"])
print(marks)
```

## loc and iloc

```python
data = pd.Series(['a' , 'b' , 'c' , 'd'], index = [101,102,103,104])
print(data)
print("-----------------------")
print('loc values') # manually generated index values
print(data.loc[101])
print(data.loc[102])
print(data.loc[103])
print(data.loc[104])

#iloc
print('iloc values') # original index values
print(data.iloc[0])
print(data.iloc[1])
print(data.iloc[2])
print(data.iloc[3])
print("------------------------------")
print(data.loc[102:104])
print("------------------------------")
print(data.iloc[1:4])
```

```python
data = {
    'Name':['A' , 'B' , 'C' , 'D' , 'E'],
    "Age":[23,25,22,24,26],
    "Dept":['HR' , 'IT' , 'FINANCE', 'IT' , 'HR'],
    "Salary":[30000,35000,28000,40000,45000]
}

df = pd.DataFrame(data, index = ['E1' ,'E2' ,'E3', 'E4','E5'])
print(df.loc['E1'])
print("----------------------------")
print(df.loc['E2'])
print("----------------------------")
print(df.loc[['E1' , 'E3']])
print("----------------------------")
print(df.loc['E2' : 'E4'])
print("----------------------------")
print(df.loc[['E2' ,'E3', 'E4'] , ['Name' , 'Salary']])
print("----------------------------")
print(df.loc[df['Salary'] > 35000])
print('iloc examples')
print("----------------------------")
print(df.iloc[0])
print("----------------------------")
print(df.iloc[[0,3]])
print("----------------------------")
print(df.iloc[0:4])
print("----------------------------")
print(df.iloc[[0,2],[0,3]]) #0,2 ---> e1,e3rows, ----> name,salary(coloumns)
print("----------------------------")
high_salary = df['Salary'] > 35000
print(df.iloc[high_salary.values])
```

## DataFrame Operations: Column & row Manipulation

```python
data = {
    'Name':["A","B","C","D","E"],
    'Age':[20,21,19,22,26],
    'Marks':[85,92,78,90,88],
    'City':["Delhi","Mumbai","Chennai" ,"Banglore" ,"Hyderabad"]
}
df = pd.DataFrame(data)
df
```

```python
# column selection

print(df['Name'])
print("--------------------------")
print(df[['Name' , 'Marks']])
```

```python
# Add new Column
df['Grade'] = ["B", "A","C","A+","B+"]
df
```

```python
#sorting
print(df.sort_values(by = 'Age'))  #ascending
print("------------------------------")
print(df.sort_values(by = 'Age' , ascending = False))
```

```python
# column deletion

df  = df.drop("City" , axis = 1)
df
```

```python
#Row selection
print(df.iloc[2])
print("-------------------------")
print(df.loc[2])
```

```python
# Add new Row
new_row = {
    "Name":"F" ,"Age":23,"Marks":92,"Grade":"A"
}
df.loc[5] = new_row
print(df)
```

```python
#Row deletion
df = df.drop(5,axis = 0)
df
```

```python
# Gives statistical values for numerical columns
df.describe()
print("-------------------------------")
print(df.describe(include = 'all'))  # for all columns
```
