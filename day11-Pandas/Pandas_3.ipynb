## DataFrame iteration Methods

## itertows()

```python
data = {
    "ID":[1,2,3],
    "Name":["A" , "B" , "C"],
    "Marks":[85,90,75]
}
df = pd.DataFrame(data)

for index , row in df.iterrows():
    print(f"Index: {index}, ID: {row["ID"]}, Name: {row["Name"]} , Marks:{row["Marks"]}")
```

## itertuples()

```python
data = {
    "ID":[1,2,3],
    "Name":["A" , "B" , "C"],
    "Marks":[85,90,75]
}
df = pd.DataFrame(data)

for  row in df.itertuples():
    print(f"Index: {index}, ID: {row.ID}, Name: {row.Name} , Marks:{row.Marks}")
```

## item()

```python
#iterate over columns of a DataFrame
#for each column it returns (column_name , Series) pair
data = {
    "ID":[1,2,3],
    "Name":["A" , "B" , "C"],
    "Marks":[85,90,75]
}
df = pd.DataFrame(data)

for col_name, col_data in df.items():
    print(f"Column Name:,{col_name}")
    print(col_data)
    print("-----------------")
```
