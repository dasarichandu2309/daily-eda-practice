### Selecting and Swapping columns

```python
df = pd.DataFrame({
    "Name":["A","B","C"],
    "Age":[23,25,21],
    "City":["Delhi" , "Mumbai" , "Chennai"]
})
df
```

```python
print(df["Name"])
print("---------------------")
print(df[["Name" , "City"]])
print("-----------------------")
print(df.iloc[:,0]) # first column
print("------------------------")
print(df.iloc[:,[0,2]])
print("-------------------------")
print(df.iloc[:,0:2])
```

```python
#Swapping
df1 = df[["Name", "Age" , "City"]]
print(df1)
print("------------------------")
#SWap two columns (2nd Method)
cols = list(df.columns)
cols[1] , cols[2]  = cols[2] , cols[1]
df2 = df[cols]
print(df2)
print("------------------------")


#3rd method ---> 
df3 = df.reindex(columns = ["City", "Name" , "Age"])
print(df3)
print("------------------------")
# 4TH method
df4 = df.iloc[:,[2,0,1]]
print(df4)
```

### Addding new column and drop existing column

```python
df = pd.DataFrame({
    "Name":["A","B","C"],
    "Age":[23,25,21],
    "City":["Delhi" , "Mumbai" , "Chennai"]
})
df
```

```python
#Add new column
df["Country"] = "India"
df
```

```python
df["Marks"] = [85,90,94]
df
df["Age_after_5years"] = df["Age"] + 5
df
```

```python
df = df.assign(Status = ["Pass" , "Pass" , "Pass"])
df
```

```python
#Dropping columns
df = df.drop("Marks" , axis = 1)
df
```

```python
df.drop("Status" , axis =1 , inplace = True) # inplace modifies the original directly , no need to reassign
df
```

### Vectorized Arthemtic Operations in pandas

```python
df = pd.DataFrame({
    "A":[10,20,30],
    "B":[1,2,3]
})
df
```

```python
df["A + B"] = df["A"] + df["B"]
df
```

```python
df["A - B"] = df["A"] - df["B"]
df
```

```python
df["A * B"] = df["A"] * df["B"]
df
```

```python
df["A / B"] = df["A"] / df["B"]
df
```

```python
df["A % B"] = df["A"] % df["B"]
df
```

```python
df["A ** B"] = df["A"] ** df["B"]
df
```

```python
df["A + 5"] = df["A"] + 5
df["B * 10"] = df["B"] * 10
df
```
