## Data Type Conversion

### astype()

```python
import pandas as pd
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame({
    "Age":["23" , "25" , "21"],
    "Salary":[1000.5, 2000.0, 1500.75],
    "Employed":[1,0,1]
})
df.info()
```

```python
# Convert age from object to int
df["Age"] = df["Age"].astype(int)
df.dtypes
```

```python
#Convert Salary from float to int
df["Salary"] = df["Salary"].astype(int)
df.info()
```

```python
df["Employed"] = df["Employed"].astype(bool)
df.info()
```

### pd.to_numeric

```python
df["Age"] = df["Age"].astype(str)
df.dtypes
df["Age"] = pd.to_numeric(df["Age"] , errors = "coerce")
print(df.dtypes)
```

### pd.to_datetime

```python
df["JoinDate"] = pd.to_datetime(['2025-09-09','2024-01-01','2023-12-15'])
df.dtypes
```

### pd.to_timedelta

```python
df["Experience"] = pd.to_timedelta(["365 days" , "730 days", "1095 days"])
df.dtypes
```

```python
df["Department"] = pd.Series(["HR" , "IT", "Finance"]).astype("category")
df.dtypes
df
```

## Boolean Row Selction by vlaues

```python
data = ({
    "Name":["A","B","C","D","E","F"],
    "Age":[25,30,35,40,45,50],
    "City":["New york" , "Paris","London","Berlin" ,"Paris" , "London"],
    "Salary":[50000,60000,70000,80000,90000,100000]
})
df = pd.DataFrame(data)
df
        
```

```python
df[df["Age"] == 30]
```

```python
df[df["City"] != 'Paris']
```

```python
df[df["Age"] >= 40]
```

```python
df[df["Age"] <= 40]
```

```python
# AND condition (&)

df[(df["Age"] > 30 )& (df["City"] == "Paris")]
```

```python
# OR Condition (|)
df[(df["City"] == "Paris" )| (df["Age"] < 30)]
```

```python
# isin() condition
df[df["City"].isin(["Paris","Hyderabad"])]
```

```python
# not in (~ with isin())
df[~df["City"].isin(["Paris","Hyderabad"])]  #give other than paris and hyderabad
```

```python
# str.contains
df[df["City"].str.contains("on")]
```

```python
# str.notcontains
df[~df["City"].str.contains("on")]
```

```python
#Between condition
df[df["Age"].between(30,45)]
```

```python
# NotBetween condition
df[~df["Age"].between(30,45)]
```

```python
# StartsWith
df[df["City"].str.startswith("N")]
```

```python
# Not StartsWith
df[~df["City"].str.startswith("N")]
```

```python
# Endswith()
df[df["City"].str.endswith("n")]
```

```python
# NOT Endswith()
df[~df["City"].str.endswith("n")]
```

```python
# age>30, city = london , salary>=7000
df[(df["Age"]>30) & (df["City"] == "London") & (df["Salary"] > 70000)]
```
