```python
import pandas as pd
```

```python
data={
    'Region':['East','East','East','East','west','west','west','west','South','South','South','South'],
    'Product':['Laptop','Laptop','Moblie','Mobile','Laptop','Laptop','Tablet','Tablet','Moblie','Mobile','Tablet','Tablet'],
    'Sales':[800,500,600,400,1000,700,500,300,650,550,450,350],
    'Quantity':[5,3,4,2,6,4,3,2,5,3,4,2]
}
df=pd.DataFrame(data)
df
```

## Crosstab

```python
data = {
    "Employee":["John","James","David","Mary","Tom","Charlie"],
    "Department":["Sales","HR","Sales","IT","HR","IT"],
    "Gender":["Male","Female","Male","Female","Male","Male"],
    "Salary":[50000,60000,55000,70000,62000,68000]
}
df = pd.DataFrame(data)
df
```

```python
# count of employees per department X gender

ct = pd.crosstab( index = df["Department"],columns = df["Gender"])
ct
```

```python
ct_sum = pd.crosstab(index = df["Department"],columns = df["Gender"],values = df["Salary"],aggfunc = "sum")
ct_sum
```

```python
ct_multi = pd.crosstab(index = df["Department"],columns = df["Gender"],values = df["Salary"],aggfunc = ["sum","mean"])
ct_multi
```

```python
ct_multi = pd.crosstab(index = df["Department"],columns = df["Gender"],values = df["Salary"],aggfunc = ["sum","mean"],margins = True)
ct_multi
```

```python
ct_norm = pd.crosstab(index = df["Department"],columns = df["Gender"],normalize = "index")
ct_norm
```

## Concatenation

```python
data_a = {
    "EmpId":[1,2],
    "Name":["John","Tom"],
    "Department":["Sales","HR"]
}

data_b = {
    "EmpId":[3,4],
    "Name":["David","Mary"],
    "Department":["IT","Sales"]
}
df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)
print(df_a)
print("--------------------------")
print(df_b)
```

```python
# Row Wise concatination

row_concat = pd.concat([df_a,df_b],axis = 0,ignore_index = True)
row_concat
```

```python
# column wise

data_a = {
    "EmpId":[1,2],
    "Name":["John","Tom"]
}

data_b = {
    "Department":["IT","Sales"],
    "Salary":[50000,60000]
}
df_a = pd.DataFrame(data_a)
df_b = pd.DataFrame(data_b)
print(df_a)
print("--------------------------")
print(df_b)
```

```python
col_concat = pd.concat([df_a,df_b] , axis = 1)
col_concat
```

## Merging

```python
employees = pd.DataFrame({
    "EmpID":[1,2,3,4],
    "EmpName":["A","B","C","D"],
    "DeptID":[101,102,103,104]
})
departments = pd.DataFrame({
    "DeptID":[101,102,103,105],
    "DeptName":["HR","IT","Fincance","Marketing"]
})

print(employees)
print("---------------------------------")
print(departments)
```

```python
# Inner Join ---> returns Common rows only
inner_join = pd.merge(employees,departments , on = "DeptID" , how  = 'inner')
print(inner_join)
```

```python
# Left Join ---> All employees + matching depts
left_join = pd.merge(employees,departments , on = "DeptID" , how  = 'left')
print(left_join)
```

```python
# Right Join ---> All Departmengts + matching employees
right_join = pd.merge(employees,departments , on = "DeptID" , how  = 'right')
print(right_join)
```

```python
# Outer Join ---> All records from both df's
outer_join = pd.merge(employees,departments , on = "DeptID" , how  = 'outer')
print(outer_join)
```
