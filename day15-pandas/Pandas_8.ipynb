## Index Based Merging

```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpName":["A","B","C","D"],
    "Dept":["HR","IT","Finance","Sales"]
               
},index = [101,102,103,104])

salaries = pd.DataFrame({
    "Salary":[50000,60000,55000,45000]
},index = [101,102,103,105])
print(employees)
print("----------------------")
print(salaries)
```

```python
#Inner Join
inner_merge = employees.merge(salaries,left_index = True,right_index = True , how = 'inner')
inner_merge
```

```python
# Left join
left_merge = employees.merge(salaries,left_index = True,right_index = True , how = 'left')
left_merge
```

```python
# Right join
right_merge = employees.merge(salaries,left_index = True,right_index = True , how = 'right')
right_merge
```

```python
# Outer join
outer_merge = employees.merge(salaries,left_index = True,right_index = True , how = 'outer')
outer_merge
```

## Data Cleaning/Handling Missing Data

```python
data = {
    "Customer":["A","B","C","D","E"],
    "Age":[25,None,30,None,28],
    "City":["New York","Los Angels",None,"Chicago",None],
    "Purchase":[200,None,150,300,None]
}
df = pd.DataFrame(data)
df
```

```python
# Detect Missing values
# df.isnull()True is missing values , False = Non missing values
print(df.isnull())
```

```python
# count Missing values in each column
df.isnull().sum()
```

```python
# Total missing vlaues in dataset
df.isnull().sum().sum()
```

```python
# Detect non missing vlaues
# df.notnull()False is missing values , True = Non missing values
df.notnull()
```

```python
#Count of non missing values in each column
df.notnull().sum()
```

```python
# Total non nmissing values in dataset
df.notnull().sum().sum()
```

```python
df1 = df.fillna(0)
df1
```

```python
df2 = df.fillna({"Age":df["Age"].mean(),"City":"Unknown","Purchase":df["Purchase"].median()})
df2
```
