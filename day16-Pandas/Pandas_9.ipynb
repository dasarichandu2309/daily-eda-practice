## String Manipulations/Operations

```python
import pandas as pd
```

```python
data = {
    "Name":["Ali    ","Bob","Charlie","David"],
    "Email":['ali@gmail.com','bob@gmail.com','charlie@gmail.com','david@gmail.com']
}
df = pd.DataFrame(data)
df
```

```python
#Changing case
print(df['Name'].str.lower())
print("---------------------------")
print(df["Name"].str.upper())
print("----------------------------")
print(df["Name"].str.title())
print("----------------------------")
print(df["Name"].str.strip())

```

```python
#slcing
df["Name"].str.strip().str[:2]
```

```python
df["Email"].str.split('@').str[1]
```

```python
df["Email"].str.contains('gmail') # check substring
```

```python
df["Name"].str.strip().str.startswith("A")
```

```python
df["Name"].str.strip().str.endswith("e")
```

```python
# Replacing
df["Email"].str.replace("gmail","yahoo")
```

```python
# Splitting and joining
df["Email"].str.split("@") # Split text
```

```python
#joining
df["Name"].str.strip().str.cat(sep = '-')
```

```python
df["Name"].str.len()
```

## Date time Manipulation

```python
data = {
    "Student":["Ali","Bob","Charlie","David"],
    "JoinDate":["2023-01-15","2023-03-22","2023-05-10","2023-07-01"],
    "Lastlogin":["2023-09-01 10:30:00","2023-09-05 15:45:00","2023-09-07 09:15:00","2023-09-10 20:00:00"]
}
df = pd.DataFrame(data)
df
```

```python
df["JoinDate"] = pd.to_datetime(df["JoinDate"])
df["Lastlogin"] = pd.to_datetime(df["Lastlogin"])
df.info()
```

```python
#Extract year month day
print(df["JoinDate"].dt.year) #print year
print("---------------------------")
print(df["JoinDate"].dt.month) #
print("----------------------------")
print(df["JoinDate"].dt.day)
print("----------------------------")
print(df["JoinDate"].dt.weekday)
print("-----------------------------")
print(df["JoinDate"].dt.day_name())
```

```python
#difference between dates
df["Lastlogin"] - df["JoinDate"]
```

```python
(df["Lastlogin"] - df["JoinDate"]).dt.days
```

```python
print(df["JoinDate"] + pd.Timedelta(days = 30))
print("----------------------------")
print(df["JoinDate"] - pd.Timedelta(days = 30))
```

```python
print(df["Lastlogin"].dt.hour)
print("-------------------")
print(df["Lastlogin"].dt.minute)
print("---------------------")
print(df["Lastlogin"].dt.second)
```

```python
df["JoinDate"].dt.strftime("%Y/%m/%d")
```

```python
df[df["JoinDate"] > "2023-03-01"]
```
