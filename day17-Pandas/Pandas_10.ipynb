## apply()

```python
data = {
    "Name":["chandu","prasad","rohith","vinod"],
    "Department":["HR","IT","Finance","IT"],
    "Salary":[50000,60000,55000,70000],
    "Joining_Year":[2019,2020,2018,2021]
}
df = pd.DataFrame(data)
df
```

```python
df["Updated_Salaries"] = df["Salary"].apply(lambda x:x+5000)
df
```

```python
def salary_category(salary):
    if salary >= 60000:
        return "High"
    else:
        return "Low"
df["Salary_Category"] = df["Salary"].apply(salary_category)
df
```

```python
import datetime as dt
dt.datetime.now()
```

```python
current_year = dt.datetime.now().year
df["Experience"] = df.apply(lambda row : current_year - row["Joining_Year"],axis = 1)
df
```

```python
df["Name_upper"] = df["Name"].apply(str.upper)
df
```

```python
df["Dept_len"] = df["Department"].str.len()
df
```

```python
def added_salary(Salary):
    Salary+= 5000
    return Salary


df["New_salary"] = df["Salary"].apply(added_salary)
df
```

```python
data = {
    "Name":["chandu","prasad","rohith","vinod"],
    "Department":["HR","IT","Finance","IT"],
    "Salary":[50000,60000,55000,70000],
    "Joining_Year":[2019,2020,2018,2021]
}
df = pd.DataFrame(data)
df
```

```python
df["Updated_salary"] = df["Salary"].map(lambda x: x+5000)
df
```

```python
# map departmnet names to codes

dept_map = {"HR":"D1","IT":"D2","Finance":"D3"}
df["Dept_code"] = df["Department"].map(dept_map)
df
```

```python
bonus = pd.Series([500,600,550,700],index = [0,1,2,3])
df["Bonus"] = df.index.map(bonus)
df
```

```python
data = {
    "Name":["chandu","prasad","rohith","vinod"],
    "Department":["HR","IT","Finance","IT"],
    "Salary":[50000,60000,55000,70000],
    "Bonus":[5000,6000,5500,6000]
}
df = pd.DataFrame(data)
df
```

```python
def add_numeric(x):
    if isinstance(x,(int , float)):
        return x + 1000
    return x

df2 = df.applymap(add_numeric)
df2
```

```python
def to_upper(x):
    if isinstance(x,str):
        return x.upper()
    return x


df3 = df.applymap(to_upper)
df3

```

```python
def high(x):
    if isinstance(x,(int,float)) and x > 60000:
        if x >7000:
            return "High"
    return x

df4 = df.applymap(high)
df4
```

```python
df_upper = df.applymap(lambda x : x.upper() if isinstance(x,str) else x)
df_upper
```

## Rolling() and Expanding()

```python
sales = pd.Series([10,20,30,40,50])
sales
```

```python
print("Rolling:\n", sales.rolling(3).mean())
print("----------------------------------")
print("Expanding:\n", sales.expanding().mean())
```
