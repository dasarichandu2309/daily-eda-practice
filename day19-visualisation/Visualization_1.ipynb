```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import warnings 
warnings.filterwarnings('ignore')
```

## Line Plot

```python
x = ["Jan","Feb","Mar","Apr","May"]
y = [400,500,600,450,900]

plt.plot(x,y,marker = 'o',linestyle = '-',color = 'b',label = 'Line 1')
plt.title("Line plot Example")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.grid(True)
plt.legend()
plt.show()
```

## Area Plot

```python
x = ["Jan","Feb","Mar","Apr","May"]
y1 = [400,500,600,450,900]
y2 = [100,100,100,100,100]

plt.fill_between(x,y1,y2,color = 'skyblue',alpha = 0.8,label = 'Area plot')
plt.title("Area  plot Example")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.show()
```

## Bar Plot

```python
x = ["Laptop","Mobile","Earphones","Tablets","Headphones"]
y = [500,300,600,250,100]

plt.bar(x,y,color = 'b',alpha = 0.6)
plt.title("Bar Plot Example")
plt.xlabel("Electronic Gadgets")
plt.ylabel("Sales")
plt.grid(axis = 'y')
plt.show()
```

## Scatter Plot

```python
x = [1,2,3,4,5]
y = [10,14,18,16,12]

plt.scatter(x,y,marker = 'o',color = 'r',label = 'Data Points')
plt.title("Line plot Example")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.legend(loc = 'upper left')
plt.show()
```

## Pie Chart

```python
x = ["Laptop","Mobile","Earphones","Tablets","Headphones"]
y = [20,15,25,30,10] # percentages

plt.pie(y, labels = x, autopct = '%1.1f%%', startangle = 90 , colors = ['gold','lightgreen','skyblue'])
plt.title("pie char example")
plt.axis('equal')
plt.show()
```

```python
x = ["Laptop","Mobile","Earphones","Tablets","Headphones"]
y = [20,15,25,30,10] # percentages

explode = (0,0.2,0,0,0)
# 0---> slice stays in place
# >0---> slice moves outward

plt.pie(y, labels = x, autopct = '%1.1f%%', startangle = 90 , colors = ['gold','lightgreen','skyblue'],explode = explode,shadow = True)
plt.title("pie char example",fontsize = 14, fontweight = 'bold')
plt.axis('equal')
plt.show()
```

## Histogram

```python
data = [10,15,15,20,20,20,25,25,30]

plt.hist(data,bins = 5,color = 'b',alpha = 0.7)
plt.title("Histogram Example")
plt.xlabel("value")
plt.ylabel("Frequency")
plt.grid(axis = 'y')
plt.legend()
plt.show()
```
