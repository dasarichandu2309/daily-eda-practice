## Working with Csv

```python
df = pd.read_csv("Data.csv")
df
```

```python
x = df["Name"]
y = df["Age"]
plt.figure(figsize = (5,3))  #Width , Height
plt.plot(x,y,marker = 'o',linestyle = '-',color = 'b')
plt.title("Line plot Example")
plt.xlabel("Name")
plt.ylabel("Age")
plt.grid(True)
plt.xticks(rotation = 45)
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (5,3))
plt.fill_between(df["Name"],df["Salary"],color = 'skyblue',alpha = 0.8)
plt.plot(df["Name"],df["Salary"],marker = 'o',linestyle = '-',color = 'r')
plt.title("Salary of employees")
plt.xlabel("Employee")
plt.ylabel("Salary")
plt.show()
```

```python
plt.figure(figsize = (6,3))
plt.bar(df["Name"],df["Salary"],color = 'y',alpha = 0.6)
plt.title("Bar Plot Example")
plt.xlabel("Name")
plt.ylabel("Salary")
plt.xticks(rotation = 34)
plt.grid(axis = 'y')
plt.show()
```

```python
plt.scatter(df["Salary"],df["Rating"],marker = 'o',color = 'r',label = 'Salary')
plt.title("Salary VS Rating")
plt.xlabel("Salary")
plt.ylabel("Rating")
plt.grid(True)
#plt.legend(loc = 'upper left')
plt.show()
```

```python
plt.figure(figsize = (6,3))
plt.hist(df["Age"],bins = 3,color = 'purple',alpha = 0.7)
plt.title("Histogram - Age Distribution")
plt.xlabel("value")
plt.ylabel("Frequency")
plt.grid(axis = 'y')
plt.legend()
plt.show()
```

```python
plt.pie(df["Salary"], labels = df["Name"], autopct = '%1.1f%%', startangle = 90 )
plt.title("pie char - Salary Distribution")
plt.axis('equal')
plt.show()
```

```python
explode = [0.1] + [0] * (len(df["Salary"]) -1)
explode = [0] + [0] + [0.1] + [0.1] + [0] + [0]

plt.pie(df["Salary"], labels = df["Name"], autopct = '%1.1f%%', startangle = 90, explode = explode,shadow = True,colors = plt.cm.Set3.colors )
# plt.cm ---> MAtplotlib's colormap mpdule
# Set3 --> categorical colormap ---> contains bright and visually distinct colors
# .color --> list of colors
plt.title("pie char - Salary Distribution by Name",fontsize = 14,fontweight = 'bold')
plt.axis('equal')
plt.show()
```

# Seaborn

```python
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.shape
```

```python
# Univariate plot --> We need to take only an axis as input(x or y)
```

```python
# Histogram
sns.histplot(data = tips, x = 'total_bill', bins = 20,color = 'b',kde = True)
# Kde ----> Kernal density estimation
plt.title("Histogram of total bill")
plt.show()
```

```python
sns.histplot(data = tips,
             x = 'total_bill',
             bins = 20,
             color = 'skyblue',
            edgecolor = 'r',
            kde = True) # Bar color
plt.xlim(0,60)
plt.ylim(0,50)

plt.title("Histogram of total bill",fontsize = 15)
plt.xlabel("Total_Bill", fontsize = 12)
plt.ylabel("Frequency",fontsize = 12)
plt.grid(True,linestyle = '--',alpha = 0.5)
plt.show()
```

## KDE plot

```python
sns.kdeplot(data = tips,x = 'total_bill',fill = True,color = 'skyblue',linewidth = 2,linestyle = '--')
plt.xlim(0,60)
plt.ylim(0,0.06)
plt.title("KDE plot of total bill")
plt.xlabel("Total Bill",fontsize = 12)
plt.ylabel("Density")
plt.grid(True, linestyle = '--')
plt.show()
```

## Box Plot

```python
sns.boxplot(x = 'day',y = 'total_bill', data  = tips)
plt.title("Box plot of total bill")
plt.show()
```
