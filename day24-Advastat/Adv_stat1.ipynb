```python
# Tossing a coin
outcomes = ["Head","Tail"]
favourable_outcomes = ["Head"]  # getting head
# probability ---> Favourabl_event / total
p_head = len(favourable_outcomes)/len(outcomes)
print("Probability of getting Head:",p_head)
```

```python
# RE,Trail,SS,Event

# Random Experiment : rolling a dice
import random
dice_roll = random.randint(1,6)
print("Dice roll Outcome:",dice_roll)

# sample space
sample_space = [1,2,3,4,5,6]
print("Sample Space:",sample_space)


# Event : Getting even number
event_even = [2,4,6]
p_even = len(event_even)/len(sample_space)

print("Probability of even number:",p_even)
```

```python
# Simple event

outcomes = ["Head","Tail"]
#Simple Event : getting head
# probability ---> Favourabl_event / total
p_simple_event = 1/len(outcomes)
print("Probability of getting Head:",p_simple_event)
```

```python
# Compount
import itertools  # Geneate all possible combinations
coins = ["Head","Tail"]

sample_space = list(itertools.product(coins ,repeat = 2))
print("Sample Space:",sample_space)

#Comound Event : Getting atleast one Head
compound_event = [i for i in sample_space if "Head" in i]
print("Getting atleast one Head :",compound_event)

# Probability of compound event
p_compound = len(compound_event)/len(sample_space)
print("Probability of getting atleast one Head:",p_compound)
```

```python
# Certain event
coin = ["Head","Tail"]
p_certain = ["Head","Tail"]

p_certain = len(p_certain)/len(coin)
print("Certain event (Head or Tail):",p_certain)
```
