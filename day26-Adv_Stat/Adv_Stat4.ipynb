## Discrete Probability Distribution(PMF,CDF)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scores = [1,2,2,3,3,3,4,4,5]
unique_scores = sorted(set(scores))
print("Unique Scores:",unique_scores)


#Calculate PMF (probability of each unique score)
pmf = [scores.count(x) / len(scores) for x in unique_scores]

# Calculate CDF (cumulative sum of PMF values)
cdf = np.cumsum(pmf)

print()
print("Score | PMF  | CDF")
for x,p,f in zip(unique_scores,pmf,cdf):
    print(f"{x}    | {p:.3f} | {f:.3f}")


#Plot of pmf using baar chart
plt.bar(unique_scores , pmf, color = 'skyblue',edgecolor = "red",label = "PMF")

#plot cdf using step plot
plt.step(unique_scores,cdf,where = 'post',color = 'red',linewidth = 2,label = "CDF")

plt.title("Discreate Probability Distribution(PMF & CDF)")
plt.xlabel("Score")
plt.ylabel("Probability")
plt.ylim(0,1.05)
plt.legend()
plt.grid(True)
plt.show()
```

## Bernoulli's Experiment

```python
# prob of success
p = 0.7

# define possible values
x_values = [0,1]
probabilities = []
for x in x_values:
    prob = (p**x) * ((1-p) ** (1-x))
    probabilities.append(prob)
    print(f"p(X = {x}) = {prob:.2f}")

plt.bar(["Fail(0)","Pass (1)"],probabilities,color = ["red", "green"],edgecolor = 'black'
       )
plt.title("Bernoulli's Experiment (p = 0.7)")
plt.ylabel("Probability")
plt.ylim(0,1)
plt.grid(True,axis ='y')
plt.show()
```

## Binomial Random Variable

```python
import math
```

```python
n = 5
p = 0.7
x_values = list(range(n+1)) # possible no.of successes

def bnomial_prob(n,k,p):
    comb = math.factorial(n)/(math.factorial(k) * math.factorial(n-k))
    prob = comb * (p**k) * ((1-p) ** (n-k))
    return prob

probabilities = []
for k in x_values:
    prob = bnomial_prob(n,k,p)
    probabilities.append(prob)
    print(f"P(X = {k}) = {prob:.4f}")
```

```python
colors = ["red", "green", "blue", "orange"]
plt.bar(x_values,probabilities, color = colors,edgecolor = 'black')
plt.title("Binomial Distribution:P(X = k) for Ravi's 5 Exams")
plt.xlabel("Number of exams Passed (X)")
plt.ylabel("Probability")
plt.xticks(x_values)
plt.ylim(0,max(probabilities) + 0.05)
plt.show()
```
