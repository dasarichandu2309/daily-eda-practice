```python
# R.E ---> Rolling a dice

sample_space = {1,2,3,4,5,6}

probabilities = {outcomes: 1/6 for outcomes in sample_space}

# Non Negativity
for outcomes, prob,in probabilities.items():
    print(f"p({outcomes}) = {prob} (>=0)")

print("---------------------------------")
# Certain Axiom
total_prob = sum(probabilities.values())
print("Sum of probabilities:",total_prob)


#Additivity axiom
sample_space = {1,2,3,4,5,6}

# Each outcome in dice has prob
prob = 1/6
#Event A : rolling an even number
A = {2,4,6}

#Event B:rolling an odd number
B = {1,3,5}

#probabilities of A and B
p_A = len(A) * prob
p_B = len(B) * prob



p_union = len(A.union(B)) * prob

print(f"Probability of A = {p_A}")
print(f"Probability of A = {p_B}")
print(f"Probability of A U B = {p_union}")
```
